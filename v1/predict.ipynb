```python
import pickle
import pandas as pd
```

```python
with open("demand_forecast_model.pkl", "rb") as f:
    model = pickle.load(f)
```

```python
test_data = pd.read_csv("data/test.csv")
```

```python
test_data
```

```python
test_data[['day', 'month', 'year']] = test_data['week'].str.split("/", expand=True)

test_data = test_data.drop('week', axis=1)

test_data = test_data.drop('record_ID', axis = 1)
    
test_data = test_data.join(pd.get_dummies(test_data.store_id, prefix='store')).drop('store_id', axis=1)
test_data = test_data.join(pd.get_dummies(test_data.sku_id, prefix='item')).drop('sku_id', axis=1)
```

```python
test_data
```

```python
y_pred = model.predict(test_data)
```

```python
output = test_data.copy()
output["Predicted Demand"] = y_pred

output.to_csv("demand_forecast.csv", index=False)

print("Predictions saved.")
```
